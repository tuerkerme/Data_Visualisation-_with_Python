## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

crypto_prices_df = This type of dataset can be used to track and analyze the price changes of cryptocurrencies over time. Such a dataset can be valuable for investors, analysts, and researchers to understand the historical performance of cryptocurrency markets, predict future price movements, and develop strategies.

crypto_returns_df = This type of dataset shows the daily returns of crypto assets for specific dates. Daily returns provide information about the performance of an asset within a particular period by reflecting the price change of an asset within a day.

In this formula, "Closing Price" represents the price of an asset at the end of a specific day, while "Opening Price" represents the price of the asset at the beginning of that day.

Such daily return datasets can help investors understand daily price movements in the cryptocurrency market, track performance, and evaluate strategies.

stocks_returns_df = This dataset appears to include the daily returns of four different stocks over a specific time period. This type of dataset can be used by investors and analysts to monitor stock performance, analyze past price movements, and understand future trends.

stocks_prices_df = This dataset likely contains the daily prices of four different stocks over a specific time period. It includes the daily closing values of stocks. Such a price dataset can be used by investors and analysts to analyze the historical performance of stocks, track price movements, conduct technical analysis, and create future price forecasts.

```python
crypto_df = pd.DataFrame(data = {'allocation %':[20, 55, 5, 17, 3]}, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
```

```python
crypto_df
```

```python
plt.figure(figsize=(8,6))

explode_list= [0,0,0,0,0.3]

color_list = ["yellowgreen","gold","burlywood", "lightcoral", "lightskyblue"]


plt.pie(crypto_df["allocation %"], labels = crypto_df.index, 
        labeldistance=0.7, 
        autopct="%.1f", 
        startangle=90, 
        shadow=True,  #görüntüyü daha da güzelleştiriyor. önemlidir.
        explode= explode_list,
        pctdistance=0.3, ##valuelerin konumu belirler
        colors= color_list
       )


plt.title('Allocation Graph')

plt.legend(loc=(0))

plt.axis('equal')  # Eksenleri eşit tutar

plt.tight_layout()

plt.show()
```

## CODING TASK #2. PLOT SINGLE & MULTIPLE LINE PLOTS USING MATPLOTLIB
- **Use Pandas read_csv to read crypto_daily_prices (BTC, ETH, and LTC)**
- **Use matplotlib on the Pandas DataFrame to plot the data**

```python
df = pd.read_csv('crypto_daily_prices.csv')
df
```

```python
df.sample(15)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(df["Date"], df["BTC"], color="mediumseagreen", lw=2, label="BTC", linestyle="--")
ax.plot(df["Date"], df["ETH"], color="midnightblue", lw=3, label="ETH", linestyle="dotted")
ax.plot(df["Date"], df["ADA"], color="gold", lw=3, label="ADA", linestyle="-.")

ax.set_title("BTC-ETH-ADA daily price", fontsize=20)
ax.set_xlabel("Tarih", fontsize=14)
ax.set_ylabel("Fiyat", fontsize=14)

tick_interval = 365
plt.xticks(df["Date"][::tick_interval], rotation=45)

for stock_symbol in ["BTC", "ETH", "ADA"]:
    max_value_date = df.loc[df[stock_symbol].idxmax()]["Date"]
    min_value_date = df.loc[df[stock_symbol].idxmin()]["Date"]
    max_value = df.loc[df[stock_symbol].idxmax()][stock_symbol]
    min_value = df.loc[df[stock_symbol].idxmin()][stock_symbol]

    ax.annotate(f'Max {stock_symbol}: {max_value:.2f}\n{max_value_date}', xy=(max_value_date, max_value),
                xytext=(max_value_date, max_value + 10), arrowprops=dict(facecolor='black', arrowstyle='->'))

    ax.annotate(f'Min {stock_symbol}: {min_value:.2f}\n{min_value_date}', xy=(min_value_date, min_value),
                xytext=(min_value_date, min_value - 10), arrowprops=dict(facecolor='black', arrowstyle='->'))

ax.legend(loc="upper left", fontsize="large")
plt.grid()
plt.show()
```

## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
- **A histogram represents data using bars with various heights**
- **Each bar groups numbers into specific ranges**
- **Taller bars show that more data falls within that specific range**
- **Plot the histogram of the returns**

```python
return_df = pd.read_csv('crypto_daily_returns.csv')
return_df
```

```python
mu = round(return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(return_df['BTC'].std(), 2) # standard deviation of distribution
 
num_bins = 40
```

```python
%matplotlib inline

fig, ax=plt.subplots()

ax.hist(return_df["ADA"], bins=num_bins, rwidth=0.8)

plt.show()
```

```python
fig, ax=plt.subplots()

ax.hist(return_df["BTC"], bins=num_bins, rwidth=0.8)
ax.hist(return_df["ETH"], bins=num_bins, rwidth=0.8);
```
