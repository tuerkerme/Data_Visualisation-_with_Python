```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Import Cancer data 
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.columns
```

```python
cancer_df
```

```python
cancer_df.target.value_counts()
```

```python
plt.figure(figsize=(10,8))

plt.figure(figsize=(10, 8))

mylabels = ["benign", "malignant"]
custom_palette = sns.color_palette(["pink", "darkblue"])


plt.pie(cancer_df['target'].value_counts(), labels=mylabels,
        colors=custom_palette,
        autopct="%.1f%%",
        startangle=90,
        shadow=True,
        pctdistance=0.3,
        textprops={'color': 'white', 'fontsize': 15}
        )

plt.title('Target')

plt.legend(loc=(0))
plt.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax= sns.countplot(data=cancer_df, x= "target")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1,p.get_height()+1))
```

Features such as mean radius, mean perimeter, mean area, worst radius, worst perimeter, and worst area represent the size of cancer cells. You can examine the effects of these characteristics on the size of cancer cells.


```python

custom_palette = sns.color_palette(["darkblue", "pink"])

fig, ax = plt.subplots(2, 3, figsize=(25, 15))

sns.scatterplot(x="mean area", y="worst area", data=cancer_df, hue=cancer_df["target"], ax=ax[0, 0], palette=custom_palette)
ax[0, 0].set_xlabel("Mean Area", fontsize=14)
ax[0, 0].set_ylabel("Worst Area", fontsize=14)

sns.scatterplot(x="mean perimeter", y="worst perimeter", data=cancer_df, hue=cancer_df["target"], ax=ax[0, 1], palette=custom_palette)
ax[0, 1].set_xlabel("Mean Perimeter", fontsize=14)
ax[0, 1].set_ylabel("Worst Perimeter", fontsize=14)

sns.scatterplot(x="mean radius", y="worst radius", data=cancer_df, hue=cancer_df["target"], ax=ax[0, 2], palette=custom_palette)
ax[0, 2].set_xlabel("Mean Radius", fontsize=14)
ax[0, 2].set_ylabel("Worst Radius", fontsize=14)

sns.scatterplot(x="mean area", y="mean perimeter", data=cancer_df, hue=cancer_df["target"], ax=ax[1, 0], palette=custom_palette)
ax[1, 0].set_xlabel("Mean Area", fontsize=14)
ax[1, 0].set_ylabel("Mean Perimeter", fontsize=14)

sns.scatterplot(x="mean area", y="mean radius", data=cancer_df, hue=cancer_df["target"], ax=ax[1, 1], palette=custom_palette)
ax[1, 1].set_xlabel("Mean Area", fontsize=14)
ax[1, 1].set_ylabel("Mean Radius", fontsize=14)

sns.scatterplot(x="mean area", y="mean concavity", data=cancer_df, hue=cancer_df["target"], ax=ax[1, 2], palette=custom_palette)
ax[1, 2].set_xlabel("Mean Area", fontsize=14)
ax[1, 2].set_ylabel("Mean Concavity", fontsize=14)

# Set the title
fig.suptitle("Analysis of the impact of features expressing size on the size of benign and malignant cancer cells", fontsize=16)

plt.show()
```

Texture and Roughness Features:

Features like mean texture, mean smoothness, worst texture, and worst smoothness describe the texture and roughness of cancer cells. These characteristics can be important in understanding the structure of cancer cells.

```python
custom_palette = sns.color_palette(["darkblue", "pink"])

fig, ax= plt.subplots(3,1, figsize=(20,20))


sns.scatterplot(x= "mean texture", y= "worst texture",data=cancer_df, hue=cancer_df["target"], ax=ax[0], palette=custom_palette)
sns.scatterplot(x= "mean smoothness", y= "worst smoothness",data=cancer_df, hue=cancer_df["target"], ax=ax[1],  palette=custom_palette)
sns.scatterplot(x= "mean texture", y= "mean smoothness",data=cancer_df, hue=cancer_df["target"], ax=ax[2],  palette=custom_palette)

fig.suptitle("Analysis according to texture and roughness", fontsize=16)
plt.show()
```

Error Features:

Features such as radius error, texture error, perimeter error, and area error represent measurement errors, and examining the impact of these errors on cancer diagnosis can be important.

When measuring benign cancer cells, the margin of error tends to be lower, but for malignant cancer cells, the margin of error is higher. This implies that a larger margin of error may increase the likelihood of the cancer being malignant.

```python
custom_palette = sns.color_palette(["darkblue", "pink"])

fig, ax= plt.subplots(2,2, figsize=(20,20))

sns.scatterplot(x= "radius error", y= "area error",data=cancer_df, hue=cancer_df["target"], ax=ax[0,0])
ax[0, 0].set_title("area X radius error")

sns.scatterplot(x= "texture error", y= "perimeter error",data=cancer_df, hue=cancer_df["target"], ax=ax[0,1])
ax[0, 1].set_title("texture X perimeter error")

sns.histplot(data=cancer_df, x="radius error", hue="target", ax=ax[1, 0], kde=True, palette=custom_palette)
ax[1, 0].set_title("Histogram: Radius Error")
sns.histplot(data=cancer_df, x="area error", hue="target", ax=ax[1, 1], kde=True, palette=custom_palette)
ax[1, 1].set_title("Histogram: Area Error")

fig.suptitle("The impact of measurement errors in cancer diagnosis", fontsize=16)
plt.show()
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
custom_palette = sns.color_palette(["darkblue", "pink"])


sns.pairplot(
    cancer_df,
    vars=["mean area", "mean radius", "mean smoothness", "mean texture"],
    corner=True,
    hue="target",
    palette=custom_palette
)
plt.show()
```
